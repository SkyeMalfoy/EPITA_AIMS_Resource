```python
import pandas as pd
import numpy as np
import random
```

```python
#Build the data set (list)
data = open("iris.data", "r")
matrix = []

for line in data:
    line = line.strip()
#     print(line)
    data_line = line.split(",")[:4]
#     print(data_line)
    float_vector = []
    if len(data_line) > 3:
        for element in data_line:
            float_vector.append(float(element))
#         print(float_vector)
        matrix.append(float_vector)
length = len(matrix[0])
```

```python
#Euclidian function
#Calculate the distance between two points on the four-dimensional space
def euclidian(p1, p2): #Input variable: two points
    if len(p1) != len(p2):
        return -1
    
    sum_distance = 0
    
    for axis in range(length):
        sum_distance += np.square(p1[axis] - p2[axis])

    return np.sqrt(sum_distance)

#test for function euclidian
# distance = euclidian(matrix[0], matrix[1])
# print(distance)
```

```python
#Generate random points from the matrix (in this case, generate 4 centroids)
k = 4
centroids = []
for _ in range(k):
    point = random.choice(matrix)
    centroids.append(point)
centroids
```

```python
#Divide the whole dataset into k clusters
def new_clusters(k, matrix, centroids):
    clusters = []
    for num in range(k):
        clusters.append([])

    for point in matrix:
        best_distance = -1
        best_index = -1
        for index in range(k):
#             print(centroids)
            euclidian_distance = euclidian(centroids[index], point)
    #         print(euclidian_distance)

            if best_distance == -1:
                best_distance = euclidian_distance
                best_index = index
            elif best_distance > euclidian_distance:
                best_distance = euclidian_distance
                best_index = index
        clusters[best_index].append(point)
    return clusters
```

```python
#Find new centroids based on the old cluster
#The average of n points in the old cluster
def centroid(cluster_single, length):
    returned_centroid = [0.0] * length
    
    #step 1: calculate the sum
    for point in cluster_single:
        for index in range(length):
            returned_centroid[index] += point[index]
#     print(returned_centroid)
#   #step 2 calculate the mean
    for index in range(length):
        returned_centroid[index] = returned_centroid[index] / len(cluster_single)
        
    return returned_centroid

def new_centroid(clusters, length):
    returned_centroid = []
    for cluster_single in clusters:

        returned_centroid.append(centroid(cluster_single, length))
    
    return returned_centroid


#test for these two functions
# new_centroids = new_centroid(clusters, length)
# new_centroids
```
