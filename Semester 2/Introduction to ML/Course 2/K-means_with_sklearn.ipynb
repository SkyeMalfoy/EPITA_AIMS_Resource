```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
matrix = pd.read_csv('iris.data', usecols = [0, 1, 2, 3])
matrix
```

```python
for k in range(2,11):
    kmeans = KMeans(n_clusters = k).fit(matrix)
    print('k={0}'.format(k))
    print(kmeans.cluster_centers_)
    print(kmeans.labels_)
```
