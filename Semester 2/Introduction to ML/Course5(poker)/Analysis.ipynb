```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df_train = pd.read_csv("new_train.data", names=['C1', 'S1', 'C2', 'S2', 'C3', 'S3', 'C4', 'S4', 'C5', 'S5', 'CLASS'])
df_test = pd.read_csv("poker-hand-testing-short.data", names=['C1', 'S1', 'C2', 'S2', 'C3', 'S3', 'C4', 'S4', 'C5', 'S5', 'CLASS'])
df_train
```

```python
#Get data for training and testing
shuffled_train = df_train.sample(frac=1)
shuffled_test = df_test.sample(frac=1)
train_x, train_y, test_x, test_y = shuffled_train.iloc[:, :-2], shuffled_train.iloc[:, -1], shuffled_test.iloc[:, :-2], shuffled_test.iloc[:, -1]
# print(train_x.shape, train_y.shape, test_x.shape, test_y.shape)
print(type(shuffled_train))
```

```python
#Make random forest 
clf = RandomForestClassifier(random_state=0)
clf.fit(train_x, train_y)

prediction = clf.predict(test_x)
print(prediction)
# print(test_y)
# clf.score(test_y, prediction)
```

```python
#Calculate the accuracy of the prediction
length = len(shuffled_test)
sum_poker = 0
for index in range(length):
    sum_poker = sum_poker + abs(prediction[index] - test_y[index])
accuracy = sum_poker / length
print(length)
print(accuracy)
```

```python
print(classification_report(prediction, test_y))
print(confusion_matrix(prediction, test_y))
```
