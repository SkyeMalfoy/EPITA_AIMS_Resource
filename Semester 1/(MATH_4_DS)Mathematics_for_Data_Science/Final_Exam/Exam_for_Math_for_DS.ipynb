```python
#5. Check the type of the variables
print(df.dtypes)
#Change the type of qualitative variables to catogery
#(column 1,2,5,6,8,10,11,12,13)['sex','cp','fbs','restecg','exang','slope','ca','thal',the_predicted_attribute']
df['sex'] = df['sex'].astype('category')
df['cp'] = df['cp'].astype('category')
df['fbs'] = df['fbs'].astype('category')
df['restecg'] = df['restecg'].astype('category')
df['exang'] = df['exang'].astype('category')
df['slope'] = df['slope'].astype('category')
df['ca'] = df['ca'].astype('category')
df['thal'] = df['thal'].astype('category')
df['the_predicted_attribute'] = df['the_predicted_attribute'].astype('category')
```

```python
#7. Apply a solution for the nan values
#1) Replace the missing value by the average value.
#E.g.
df["age"] = df["age"].fillna(df["age"].mean())

before_del = pd.DataFrame(df)
print(before_del)
#2) Delete the column is there are too much Nan values in certain column.
#E.g.
del df['age']
after_del = pd.DataFrame(df)
print(after_del)
#We can see from the result that: "Column 0" is being deleted.

#3) Delete the row if there are not too much Nan values in certain column.
df = df.drop([0])
print(pd.DataFrame(df))
#We can see from the result that: "Row 0" is being deleted.
```

```python
#11. Study the links between quantitatives variables (correlations, scatter, plot, metrics)
#Quantitative variables ((column 0,3,4,7,9)["age","trestbps",'chol','thalach','oldpeak']
corr = df[["age","trestbps",'chol','thalach','oldpeak']].corr()
#print out the correlations
print('The correlations:')
print(corr)
print('\n')
#Scatter:
plt.figure(figsize=(10,5))
plt.suptitle("Scatter plots")
plt.subplot(121)
plt.scatter(df['age'], df['trestbps'])
plt.subplot(122)
plt.scatter(df['chol'], df['thalach'])
plt.show()
print('\n')
#Plot:
plt.figure(figsize=(10,5))
plt.suptitle("Plots")
plt.subplot(121)
plt.scatter(df['age'], df['trestbps'])
plt.subplot(122)
plt.scatter(df['chol'], df['thalach'])
plt.show()
print('\n')
#Heatmap
plt.title("heatmap")
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

```python
#12. Study the link between qualitative variables (chi-squared)
#Qualitative variables(column 1,2,5,6,8,10,11,12,13) ['sex','cp','fbs','restecg','exang','slope','ca','thal','the_predicted_attribute']
from scipy.stats import chi2_contingency
def chi2(c1, c2):
    table = pd.crosstab(df[c1],df[c2])
    stat, p, dof, expected = chi2_contingency(table)
    print("This is the result between:column{0} and column{1}".format(c1, c2))
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print('Probably independent')
    else:
        print('Probably dependent')
    print()

l = ['sex','cp','fbs','restecg','exang','slope','ca','thal','the_predicted_attribute']
for i,c1 in enumerate(l):
    for c2 in l[i+1:]:
        chi2(c1,c2)
```

```python
#13.Study the link between qualitative vs quantitative variables (ANOVA)
#Here we use an example between cp(chest pain type) and age:
import os
import random
import patsy
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.stats.anova import AnovaRM
from statsmodels.regression.mixed_linear_model import MixedLMResults
from scipy import stats
# #Do cat_codes on variables:
# df['cp'] = df['cp'].cat.codes
# df['age'] = df['age'].cat.codes
# inspect data
sns.catplot(x='cp', y="age", data=df, dodge=True, kind='violin', aspect=3)
plt.title("ANOVA")
plt.show()

# generate model for linear regression
my_model = smf.ols(formula='age ~ cp', data=df)

# fit model to data to obtain parameter estimates
my_model_fit = my_model.fit()

# print summary of linear regression
print(my_model_fit.summary())

# show anova table
anova_table = sm.stats.anova_lm(my_model_fit, typ=2)
print(anova_table)
```
